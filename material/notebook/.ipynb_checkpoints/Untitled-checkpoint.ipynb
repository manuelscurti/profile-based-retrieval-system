<h1>Retrieving and processing an XML file from the Internet</h1>
Retrieving articles from NYTimes RSS feed

```python
import urllib3 as urllib
import xmltodict
import collections
```

```python
def homepage(request):
    file = urllib.urlopen('https://www.goodreads.com/review/list/20990068.xml?key=nGvCqaQ6tn9w4HNpW8kquw&v=2&shelf=toread')
    data = file.read()
    file.close()

    data = xmltodict.parse(data)
    return render_to_response('my_template.html', {'data': data})
```

```python
test = homepage("http://rss.nytimes.com/services/xml/rss/nyt/Science.xml")
```

```python
http = urllib.PoolManager()
r = http.request('GET', 'http://rss.nytimes.com/services/xml/rss/nyt/Science.xml')
```

```python
for article in data:
    article_title = ""
    article_descr = ""
    article_add_descr = ""
    
    if "title" in article:
        article_title = article["title"]
        
    if "media:description" in article:
        article_descr = article["media:description"]
    
    if "description" in article:
        article_add_descr = article["description"]
    
    if not isinstance(article_title, str):
        article_title = ""
    
    if not isinstance(article_descr, str): 
        article_descr = ""
    
    if not isinstance(article_add_descr, str): 
        article_add_descr = ""
    
    print("[debug] "+article_title)
    print(type(article_descr))
    
    if article_title != "" or article_descr != "" or article_add_descr != "":
        print("[start]"+article_title +article_descr + article_add_descr+"[end]\n")

    
```
